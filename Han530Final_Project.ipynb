```python
high_school = data[data['education'] == 'High school diploma']
some_college = data[data['education'] == 'Some college or higher']

high_school_counts = high_school['Count of employed veterans in year 1'].dropna()
some_college_counts = some_college['Count of employed veterans in year 1'].dropna()

# Calculate PMFs 
high_school_pmf = thinkstats2.Pmf(high_school_counts, label='High School Diploma')
some_college_pmf = thinkstats2.Pmf(some_college_counts, label='Some College or Higher')

thinkplot.PrePlot(2, cols=2)

# Left plot
thinkplot.Pmf(high_school_pmf, color='blue', label='High School Diploma')
thinkplot.Pmf(some_college_pmf, color='green', label='Some College or Higher')
thinkplot.Config(xlabel='Employment Count', ylabel='Probability', axis=[0, 45000, 0, 0.5], legend=True, grid=True, title='PMF of Employed Veterans in Year 1')

# Right plot
thinkplot.PrePlot(2)
thinkplot.SubPlot(2)
thinkplot.Pmfs([high_school_pmf, some_college_pmf])
thinkplot.Config(xlabel='Employment Count', ylabel='Probability', axis=[0, 45000, 0, 0.5], legend=True, grid=True, title='Step Functions of PMFs')

thinkplot.Show()
```

```python
#Filter
employed_veterans = data['Count of employed veterans in year 1'].dropna()
lambda_estimate = 1 / employed_veterans.mean()
x_vals = np.linspace(0, employed_veterans.max(), 100)

# Calculate PDF of the Exponential distribution
pdf_vals = stats.expon.pdf(x_vals, scale=1/lambda_estimate)

# Create a PMF 
employed_veterans_pmf = thinkstats2.Pmf(employed_veterans, label='Employed Veterans Year 1')

# Plot the histogram 
thinkplot.Hist(employed_veterans_pmf, align='right', width=500)

# Plot the Exponential distribution 
thinkplot.Plot(x_vals, pdf_vals, label='Fitted Exponential Distribution')
thinkplot.Config(xlabel='Employment Count', ylabel='Probability', axis=[0, 45000, 0, 0.15])
thinkplot.Show()
```

```python
# Filter 
data_clean = data.dropna(subset=['education', 'Count of employed veterans in year 1'])

education_map = {'General Educational Development (GED) test': 1,
                 'High school diploma': 2,
                 'Some college or higher': 3}
data_clean['education_numeric'] = data_clean['education'].map(education_map)

# Create first scatter plot (education vs employment)
plt.figure(figsize=(10, 5))

# Scatter plot for Education vs Employment Count
plt.subplot(1, 2, 1)
plt.scatter(data_clean['education_numeric'], data_clean['Count of employed veterans in year 1'], color='blue')
plt.title('Education vs Employment Count')
plt.xlabel('Education Level (1=GED, 2=High School, 3=Some College or Higher)')
plt.ylabel('Employment Count')
plt.xticks([1, 2, 3], ['GED', 'High School', 'Some College or Higher'])

# Create second scatter plot (Employment Count vs Non-employment Count)
plt.subplot(1, 2, 2)
plt.scatter(data_clean['Count of employed veterans in year 1'], data_clean['Count of nonemployed veterans in year 1'], color='green')
plt.title('Employment Count vs Non-Employment Count')
plt.xlabel('Employment Count')
plt.ylabel('Non-Employment Count')

# Show both plots
plt.tight_layout()
plt.show()

# Calculate Covariance and Pearson's Correlation for the first scatter plot
covariance = np.cov(data_clean['education_numeric'], data_clean['Count of employed veterans in year 1'])[0][1]
pearson_corr, _ = pearsonr(data_clean['education_numeric'], data_clean['Count of employed veterans in year 1'])

# Print
print(f"Covariance between Education and Employment Count: {covariance}")
print(f"Pearson's correlation coefficient: {pearson_corr}")

# Calculate Covariance and Pearson's Correlation for the second scatter plot
covariance2 = np.cov(data_clean['Count of employed veterans in year 1'], data_clean['Count of nonemployed veterans in year 1'])[0][1]
pearson_corr2, _ = pearsonr(data_clean['Count of employed veterans in year 1'], data_clean['Count of nonemployed veterans in year 1'])

# Print
print(f"Covariance between Employment Count and Non-Employment Count: {covariance2}")
print(f"Pearson's correlation coefficient: {pearson_corr2}")
```

```python
# Filter
high_school = data_clean[data_clean['education'] == 'High school diploma']
some_college = data_clean[data_clean['education'] == 'Some college or higher']
high_school_employment = high_school['Count of employed veterans in year 1']
some_college_employment = some_college['Count of employed veterans in year 1']
t_stat, p_value = ttest_ind(high_school_employment, some_college_employment)

#Print
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
data_clean['education_numeric'] = data_clean['education'].map({
    'High school diploma': 1,
    'Some college or higher': 2,
    'General Educational Development (GED) test': 3
})

X = data_clean['education_numeric']
y = data_clean['Count of employed veterans in year 1']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())

# Plot the regression line
plt.scatter(data_clean['education_numeric'], y, label='Data points')
plt.plot(data_clean['education_numeric'], model.fittedvalues, color='red', label='Regression line')
plt.xlabel('Education Level (Numeric)')
plt.ylabel('Employment Count (Year 1)')
plt.title('Simple Linear Regression: Education Level vs Employment Count (Year 1)')
plt.legend()
plt.show()
```
